# Part 2: Selecting Data

```python
import pandas as pd
```

```python
df = pd.read_csv('vgsales.csv')
df
```

### Sorting Rows

```python
# Sort by single column
temp_df = df.sort_values(by='Year', ascending=True)
temp_df
```

```python
# Sort by multiple columns
temp_df = df.sort_values(by=['Publisher', 'Global_Sales'], ascending=[True, False])
temp_df
```

### Subsetting Columns

```python
# Subset single column
df['Publisher']
```

```python
# Subset multiple columns
df[['Name','Publisher', 'Global_Sales']]
```

### Subsetting Rows

```python
# What games were released in 2010?
df[df['Year'] == 2010]
```

```python
# What games were published by Nintendo?
df[df['Publisher'] == 'Nintendo']
```

```python
# What games were released in 2010 and sold more than 5 million copies?
df[(df['Year'] == 2010) & (df['Global_Sales'] > 5)]
```

```python
# What games were released between 2010 and 2015 and sold more than 5 million copies?
df[(df['Year'] >= 2010) & (df['Year'] <= 2015) & (df['Global_Sales'] > 5)]
```

```python
# What games are not published by Nintendo?
df[df['Publisher'] != 'Nintendo']
```

### Subsetting Rows by categorical variables

```python
publishers = ['Riot Games','Nintendo', 'Activision',  ]
df[df['Publisher'].isin(publishers)]
```
