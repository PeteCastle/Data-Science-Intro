### Creating your own DataFrame

Suppose we have the following table:
| game_id | Name              | Genre      | Platform |
|---------|-------------------|------------|----------|
| 45      | Dota 2            | MOBA       | PC       |
| 322     | League of Legends | MOBA       | PC       |
| 32      | Mobile Legends    | MOBA       | Mobile   |
| 21      | Mario Kart        | Racing     | Switch   |
| 232     | Minecraft         | Adventure  | PS4      |

```python
df = pd.DataFrame({
    'game_id': [45, 322, 32, 21, 232],
    'Name': ['Dota 2', 'League of Legends', 'Mobile Legends', 'Mario Kart', 'Minecraft'],
    'Genre': ['MOBA', 'MOBA', 'MOBA', 'Racing', 'Adventure'],
    'Platform': ['PC', 'PC', 'Mobile', 'Switch', 'PS4']
})
df
```

### Setting DataFrame Index

```python
df.set_index('Rank', inplace=True)
df
```

### Renaming Columns

```python
columns = {
    # Key: Old Value -> Value: New Value
    "Global_Sales": "Total_Sales",
    "Publisher": "Company",
}
df.rename(columns=columns, inplace=True)
df
```

### Dropping Columns and Rows

```python
# Dropping columns
temp_df = df.drop(columns=['Total_Sales', 'Year'])
temp_df
```

```python
# Dropping rows
temp_df = df.drop(index=[1, 2, 3])
temp_df
```
